```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)

plt.rcParams["figure.figsize"] = (10,6)
```

```python
file_path = r"C:\Users\dell\Desktop\Project\data\cleaned\inc5000_cleaned.xlsx"

df = pd.read_excel(file_path)

print("Dataset Loaded Successfully")
```

## Executive KPI Summary

```python
kpis = pd.DataFrame({
    "Metric":[
        "Total Companies",
        "Total revenue",
        "Average revenue",
        "Median revenue",
        "Total employees",
        "Average employees",
        "Average Growth Rate",
        "Industries",
        "States",
        "Metro Areas"
    ],
    "Value":[
        len(df),
        df["revenue_usd"].sum(),
        df["revenue_usd"].mean(),
        df["revenue_usd"].median(),
        df["employee_count"].sum(),
        df["employee_count"].mean(),
        df["growth_percentage"].mean(),
        df["industry"].nunique(),
        df["state_name"].nunique(),
        df["metro"].nunique()
    ]
})

kpis
```

## Business Question 1

Which industries generate the highest revenue?

```python
industry_revenue = (
    df.groupby("industry", as_index=False)
      .agg(
          Total_Revenue=("revenue_usd","sum"),
          Companies=("company","count")
      )
      .sort_values("Total_Revenue", ascending=False)
)

industry_revenue.head(10)
```

## Business Question 2

State-wise company distribution

```python
state_summary = (
    df.groupby("state_name", as_index=False)
      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum")
      )
      .sort_values("Companies", ascending=False)
)

state_summary.head(10)
```

## Business Question 3

Top Metro Areas

```python
metro_summary = (
    df.groupby("metro", as_index=False)
      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum")
      )
      .sort_values("Companies", ascending=False)
)

metro_summary.head(10)
```

## Business Question 4

Top Revenue Companies

```python
top_revenue = (
    df.nlargest(
        15,
        "revenue_usd"
    )[
        [
            "company",
            "industry",
            "state_name",
            "revenue_usd"
        ]
    ]
)

top_revenue
```

## Business Question 5

Fastest Growing Companies

```python
top_growth = (
    df.nlargest(
        15,
        "growth_percentage"
    )[
        [
            "company",
            "industry",
            "state_name",
            "growth_percentage",
        ]
    ]
)

top_growth
```

## Business Question 6

Revenue Efficiency

```python
df["revenue_per_employee"] = (
    df["revenue_usd"] /
    df["employee_count"]
)

efficiency = (
    df.sort_values(
        "revenue_per_employee",
        ascending=False
    )[
        [
            "company",
            "revenue_per_employee",
            "revenue_usd",
            "employee_count"
        ]
    ]
)

efficiency.head(15)
```

## Business Question 7

Average Revenue by Industry

```python
industry_avg = (
    df.groupby("industry", as_index=False)
      .agg(
          Average_Revenue=("revenue_usd","mean"),
          Average_Employees=("employee_count","mean"),
          Average_Growth=("growth_percentage","mean")
      )
      .sort_values(
          "Average_Revenue",
          ascending=False
      )
)

industry_avg
```

## Business Question 8

Company Size Distribution

```python
company_size = (
    df.groupby("company_size", as_index=False)
      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum"),
          Employees=("employee_count","sum")
      )
)

company_size
```

## Business Question 9

Revenue Category Performance

```python
revenue_category = (
    df.groupby("revenue_category", as_index=False)
      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum")
      )
      .sort_values("revenue_category", ascending=False)
)

revenue_category
```

## Business Question 10

Growth Category Distribution

```python
growth_category = (
    df.groupby("growth_category", as_index=False)
      .agg(
          Companies=("company","count"),
          Average_Growth=("growth_percentage","mean")
      )
)

growth_category
```

## Business Question 11

Employee Size Band Analysis

```python
employee_band = (
    df.groupby("employee_size_band", as_index=False)
      .agg(
          Companies=("company","count"),
          Employees=("employee_count","sum")
      )
)

employee_band
```

## Business Question 12

Region-wise Analysis

```python
region_summary = (
    df.groupby("state region", as_index=False)
      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum"),
          Employees=("employee_count","sum")
      )
      .sort_values("Revenue", ascending=False)
)

region_summary
```

## Business Question 13

Largest Employers

```python
largest_employers = (
    df.nlargest(
        15,
        "employee_count"
    )[
        [
            "company",
            "employee_count",
            "revenue_usd"
        ]
    ]
)

largest_employers
```

## Business Question 14

Overall Company Ranking

```python
ranking = (
    df[
        [
            "rank",
            "company",
            "industry",
            "revenue_usd",
            "growth_percentage",
            "employee_count"
        ]
    ]
    .sort_values("rank")
)

ranking.head(20)
```

## Executive Summary

```python
summary = pd.DataFrame({
    "Metric":[
        "Highest Revenue Industry",
        "Top Revenue Company",
        "Fastest Growing Company",
        "Largest Employer",
        "State with Most Companies"
    ],
    "Value":[
        industry_revenue.iloc[0]["industry"],
        top_revenue.iloc[0]["company"],
        top_growth.iloc[0]["company"],
        largest_employers.iloc[0]["company"],
        state_summary.iloc[0]["state_name"]
    ]
})

summary
```
