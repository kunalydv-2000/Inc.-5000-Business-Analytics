All charts are automatically saved into the **charts/** folder.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)

plt.rcParams["figure.figsize"] = (10,6)

# Create charts folder if it doesn't exist
charts_dir = Path("../charts")
charts_dir.mkdir(parents=True, exist_ok=True)
```

```python
file_path = r"C:\Users\dell\Desktop\Project\data\cleaned\inc5000_cleaned.xlsx"

df = pd.read_excel(file_path)

print("Dataset Loaded Successfully")
```

## Revenue Distribution

```python
plt.figure()

plt.hist(df["revenue_usd"], bins=10)

plt.title("Revenue Distribution")

plt.xlabel("Revenue")

plt.ylabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "01_revenue_distribution.png", dpi=300)

plt.show()
```

## Employee Distribution

```python
plt.figure()

plt.hist(df["employee_count"], bins=10)

plt.title("Employee Distribution")

plt.xlabel("Employees")

plt.ylabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "02_employee_distribution.png", dpi=300)

plt.show()
```

## Growth Rate Distribution

```python
plt.figure()

plt.hist(df["growth_percentage"], bins=10)

plt.title("Growth Rate Distribution")

plt.xlabel("Growth Rate")

plt.ylabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "03_growth_distribution.png", dpi=300)

plt.show()
```

## Top 10 Industries

```python
industry = (
    df["industry"]
    .value_counts()
    .head(10)
)

plt.figure()

industry.sort_values().plot(kind="barh")

plt.title("Top 10 Industries")

plt.xlabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "04_top_industries.png", dpi=300)

plt.show()
```

## Top States

```python
states = (
    df["state_name"]
    .value_counts()
    .head(10)
)

plt.figure()

states.sort_values().plot(kind="barh")

plt.title("Top States")

plt.xlabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "05_top_states.png", dpi=300)

plt.show()
```

## Revenue by Industry

```python
industry_revenue = (
    df.groupby("industry")["revenue_usd"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure()

industry_revenue.sort_values().plot(kind="barh")

plt.title("Revenue by Industry")

plt.xlabel("Revenue")

plt.tight_layout()

plt.savefig(charts_dir / "06_revenue_by_industry.png", dpi=300)

plt.show()
```

```python
## Revenue by State
```

```python
state_revenue = (
    df.groupby("state_name")["revenue_usd"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure()

state_revenue.sort_values().plot(kind="barh")

plt.title("Revenue by State")

plt.xlabel("Revenue")

plt.tight_layout()

plt.savefig(charts_dir / "07_revenue_by_state.png", dpi=300)

plt.show()
```

## Revenue Category Distribution

```python
revenue_category = (
    df["revenue_category"]
    .value_counts()
)

plt.figure()

revenue_category.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Revenue Category")

plt.tight_layout()

plt.savefig(charts_dir / "08_revenue_category.png", dpi=300)

plt.show()
```

```python
## Growth Category Distribution
```

```python
growth = (
    df["growth_category"]
    .value_counts()
)

plt.figure()

growth.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Growth Category")

plt.tight_layout()

plt.savefig(charts_dir / "09_growth_category.png", dpi=300)

plt.show()
```

## Company Size

```python
company_size = (
    df["company_size"]
    .value_counts()
)

plt.figure()

company_size.plot(kind="bar")

plt.title("Company Size")

plt.xlabel("Category")

plt.ylabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "10_company_size.png", dpi=300)

plt.show()
```

## Employee Size Band

```python
employee_band = (
    df["employee_size_band"]
    .value_counts()
)

plt.figure()

employee_band.plot(kind="bar")

plt.title("Employee Size Band")

plt.tight_layout()

plt.savefig(charts_dir / "11_employee_size_band.png", dpi=300)

plt.show()
```

## Top 10 Companies by Revenue

```python
top_rev = (
    df.nlargest(10,"revenue_usd")
)

plt.figure()

plt.barh(
    top_rev["company"],
    top_rev["revenue_usd"]
)

plt.title("Top Companies by Revenue")

plt.tight_layout()

plt.savefig(charts_dir / "12_top_companies_revenue.png", dpi=300)

plt.show()
```

## Revenue vs Employees

```python
plt.figure()

plt.scatter(
    df["employee_count"],
    df["revenue_usd"],
    alpha=0.6
)

plt.title("Revenue vs Employees")

plt.xlabel("Employees")

plt.ylabel("Revenue")

plt.tight_layout()

plt.savefig(charts_dir / "13_revenue_vs_employees.png", dpi=300)

plt.show()
```

## Growth Rate vs Revenue

```python
plt.figure()

plt.scatter(
    df["growth_percentage"],
    df["revenue_usd"],
    alpha=0.6
)

plt.title("Growth Rate vs Revenue")

plt.xlabel("Growth Rate")

plt.ylabel("Revenue")

plt.tight_layout()

plt.savefig(charts_dir / "14_growth_vs_revenue.png", dpi=300)

plt.show()
```

## Correlation Heatmap

```python
corr = (
    df.select_dtypes(include="number")
      .corr()
)

plt.figure(figsize=(8,7))

plt.imshow(corr)

plt.colorbar()

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Correlation Matrix")

plt.tight_layout()

plt.savefig(charts_dir / "15_correlation_heatmap.png", dpi=300)

plt.show()
```

## Top Metro Areas

```python
metro = (
    df["metro"]
    .value_counts()
    .head(10)
)

plt.figure()

metro.sort_values().plot(kind="barh")

plt.title("Top Metro Areas")

plt.xlabel("Companies")

plt.tight_layout()

plt.savefig(charts_dir / "16_top_metro_areas.png", dpi=300)

plt.show()
```
