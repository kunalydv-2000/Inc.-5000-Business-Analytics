```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)

plt.rcParams["figure.figsize"] = (10,6)
```

```python
# Path to dataset

file_path = r"C:\Users\dell\Desktop\Project\data\cleaned\inc5000_cleaned.xlsx"

df = pd.read_excel(file_path)

print("Dataset loaded successfully.")
```

## Dataset Overview

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])

display(df.head())
```

## Descriptive Statistics

```python
df.describe().T
```

## Numerical Columns Summary

```python
numeric_df = df.select_dtypes(include="number")

summary = numeric_df.agg([
    "count",
    "mean",
    "median",
    "std",
    "min",
    "max"
]).T

summary
```

## Categorical Columns

```python
categorical = df.select_dtypes(include="object").columns

for col in categorical:
    print("="*50)
    print(col)
    display(df[col].value_counts().head(10))
```

## Revenue Distribution

```python
plt.figure()

df["revenue_usd"].hist(bins=30)

plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")

plt.show()
```

## Employee Distribution

```python
plt.figure()

df["employee_count"].hist(bins=25)

plt.title("Employee Distribution")

plt.xlabel("employee")

plt.ylabel("frequency")

plt.show()
```

## Growth Rate Distribution

```python
plt.figure()

df["growth_percentage"].hist(bins=30)

plt.title("Growth Rate Distribution")

plt.xlabel("growth_percentage")

plt.ylabel("Companies")

plt.show()
```

## Top Industries

```python
industry = (
    df["industry"]
    .value_counts()
    .head(10)
)

plt.figure()

industry.plot(kind="bar")

plt.title("Top 10 Industries")

plt.xlabel("Industry")

plt.ylabel("Companies")

plt.xticks(rotation=45)

plt.show()
```

## Top States

```python
states = (
    df["state_name"]
    .value_counts()
    .head(10)
)

plt.figure()

states.plot(kind="bar")

plt.title("Top States")

plt.xlabel("State")

plt.ylabel("Companies")

plt.xticks(rotation=45)

plt.show()
```

## Revenue by Industry

```python
industry_revenue = (
    df.groupby("industry")["revenue_usd"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure()

industry_revenue.plot(kind="bar")

plt.title("Revenue by Industry")

plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

## Employees by Industry

```python
employees = (
    df.groupby("industry")["employee_count"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure()

employees.plot(kind="bar")

plt.title("Employees by Industry")

plt.ylabel("Employees")

plt.xticks(rotation=45)

plt.show()
```

## Top Companies by Revenue

```python
top_rev = (
    df.nlargest(10,"revenue_usd")
)

display(
    top_rev[
        ["company","revenue_usd"]
    ]
)
```

```python
## Fastest Growing Companies
```

```python
top_growth = (
    df.nlargest(
        10,
        "growth_percentage"
    )
)

display(
    top_growth[
        ["company","growth_percentage"]
    ]
)
```

## Revenue per Employee

```python
df["revenue_per_employee"] = (
    df["revenue_usd"] /
    df["employee_count"]
)

display(
    df[
        ["company",
         "revenue_per_employee"]
    ]
    .sort_values(
        "revenue_per_employee",
        ascending=False
    )
    .head(10)
)
```

## Correlation Matrix

```python
corr = numeric_df.corr()

corr
```

## Correlation Heatmap

```python
plt.figure(figsize=(8,6))

plt.imshow(corr)

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.colorbar()

plt.title("Correlation Matrix")

plt.show()
```

## Revenue vs Employees

```python
plt.figure()

plt.scatter(
    df["employee_count"],
    df["revenue_usd"]
)

plt.xlabel("employee_count")

plt.ylabel("revenue_usd")

plt.title("Revenue vs Employees")

plt.show()
```

## Growth vs Revenue

```python
plt.figure()

plt.scatter(
    df["growth_percentage"],
    df["revenue_usd"]
)

plt.xlabel("growth_percentage")

plt.ylabel("revenue_usd")

plt.title("Growth vs Revenue")

plt.show()
```

## Key Business KPIs

```python
kpis = pd.DataFrame({

    "Metric":[
        "Total Companies",
        "Total Revenue",
        "Average Revenue",
        "Median Revenue",
        "Average Employees",
        "Average Growth Rate",
        "Industries",
        "States"
    ],

    "Value":[
        len(df),
        df["revenue_usd"].sum(),
        df["revenue_usd"].mean(),
        df["revenue_usd"].median(),
        df["employee_count"].mean(),
        df["growth_percentage"].mean(),
        df["industry"].nunique(),
        df["state_name"].nunique()
    ]

})

kpis
```
