```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
pd.set_option("display.float_format", "{:,.2f}".format)

print("Libraries imported successfully.")
```

## Load Dataset

```python
# Path to dataset

file_path = r"C:\Users\dell\Desktop\Project\data\cleaned\inc5000_cleaned.xlsx"

df = pd.read_excel(file_path)

print("Dataset loaded successfully.")
```

```python
df.head()
```

```python
df.tail()
```

## Dataset Dimensions

```python
rows, cols = df.shape

print(f"Rows    : {rows:,}")
print(f"Columns : {cols}")
```

## Column Names

```python
for i, column in enumerate(df.columns, start=1):
    print(f"{i:02d}. {column}")
```

## Data Types

```python
df.dtypes
```

```python
df.info
```

## Dataset Overview

```python
overview = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.values,
    "Non-Null Count": df.count().values,
    "Missing Values": df.isna().sum().values,
    "Unique Values": df.nunique().values
})

overview
```

## Numerical Columns

```python
numerical_columns = df.select_dtypes(include="number").columns

print(f"Total Numerical Columns: {len(numerical_columns)}")
print()

for col in numerical_columns:
    print(col)
```

## Categorical Columns

```python
categorical_columns = df.select_dtypes(include=["object", "category"]).columns

print(f"Total Categorical Columns: {len(categorical_columns)}")
print()

for col in categorical_columns:
    print(col)
```

## Date Columns

```python
date_columns = df.select_dtypes(include=["datetime64[ns]"]).columns

if len(date_columns) == 0:
    print("No datetime columns detected.")
else:
    print(date_columns)
```

## First Five Records

```python
df.sample(5, random_state=42)
```

## Dataset Summary

```python
summary = {
    "Rows": df.shape[0],
    "Columns": df.shape[1],
    "Numerical Columns": len(numerical_columns),
    "Categorical Columns": len(categorical_columns),
    "Missing Values": int(df.isna().sum().sum()),
    "Duplicate Rows": int(df.duplicated().sum())
}

summary_df = pd.DataFrame(summary.items(), columns=["Metric", "Value"])

summary_df
```
