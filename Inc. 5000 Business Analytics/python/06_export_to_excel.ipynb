# Notebook 06 - Export Analysis to Excel

## Objectives

This notebook exports all business summaries, KPIs, and analysis results into a single Excel workbook for reporting and dashboard creation.

### Outputs

- Executive KPIs
- Dataset Summary
- Industry Summary
- State Summary
- Metro Summary
- Revenue Category Summary
- Growth Category Summary
- Company Size Summary
- Employee Size Summary
- Top Revenue Companies
- Top Growth Companies
- Revenue Efficiency
- Overall Rankings

The exported workbook can be used directly in Excel dashboards.

````python
import pandas as pd
from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)
````

````python
file_path = r"C:\Users\dell\Desktop\Project\data\cleaned\inc5000_cleaned.xlsx"

df = pd.read_excel(file_path)

print("Dataset Loaded Successfully")
````

````python
# Create exports folder

export_folder = Path("../data/exports")

export_folder.mkdir(parents=True, exist_ok=True)
````

## Executive KPI Table

````python
kpis = pd.DataFrame({

    "Metric":[
        "Total Companies",
        "Total Revenue",
        "Average Revenue",
        "Median Revenue",
        "Maximum Revenue",
        "Minimum Revenue",
        "Total Employees",
        "Average Employees",
        "Average Growth Rate",
        "Industries",
        "States",
        "Metro Areas"
    ],

    "Value":[
        len(df),
        df["revenue_usd"].sum(),
        df["revenue_usd"].mean(),
        df["revenue_usd"].median(),
        df["revenue_usd"].max(),
        df["revenue_usd"].min(),
        df["employee_count"].sum(),
        df["employee_count"].mean(),
        df["growth_percentage"].mean(),
        df["industry"].nunique(),
        df["state_name"].nunique(),
        df["metro"].nunique()
    ]

})

kpis
````

## Dataset Summary

````python
dataset_summary = pd.DataFrame({

    "Metric":[
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicate Rows"
    ],

    "Value":[
        df.shape[0],
        df.shape[1],
        df.isna().sum().sum(),
        df.duplicated().sum()
    ]

})

dataset_summary
````

````python
industry_summary = (

    df.groupby("industry", as_index=False)

      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum"),
          Employees=("employee_count","sum"),
          Average_Revenue=("revenue_usd","mean"),
          Average_Growth=("growth_percentage","mean")
      )

      .sort_values("Revenue", ascending=False)

)
````

````python
state_summary = (

    df.groupby("state_name", as_index=False)

      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum"),
          Employees=("employee_count","sum")
      )

      .sort_values("Revenue", ascending=False)

)
````

````python
metro_summary = (

    df.groupby("metro", as_index=False)

      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum")
      )

      .sort_values("Revenue", ascending=False)

)
````

````python
revenue_category = (

    df.groupby("revenue_category", as_index=False)

      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum")
      )

)
````

````python
growth_category = (

    df.groupby("growth_category", as_index=False)

      .agg(
          Companies=("company","count"),
          Average_Growth=("growth_percentage","mean")
      )

)
````

````python
company_size = (

    df.groupby("company_size", as_index=False)

      .agg(
          Companies=("company","count"),
          Revenue=("revenue_usd","sum"),
          Employees=("employee_count","sum")
      )

)
````

````python
employee_size = (

    df.groupby("employee_size_band", as_index=False)

      .agg(
          Companies=("company","count"),
          Employees=("employee_count","sum")
      )

)
````

````python
top_revenue = (

    df.nlargest(
        25,
        "revenue_usd"
    )

)
````

````python
top_growth = (

    df.nlargest(
        25,
        "growth_percentage"
    )

)
````

````python
df["revenue_per_employee"] = (

    df["revenue_usd"] /

    df["employee_count"]

)

efficiency = (

    df.sort_values(
        "revenue_per_employee",
        ascending=False
    )

)
````

````python
ranking = (

    df.sort_values(
        "rank"
    )

)
````

## Export Workbook

````python
output_file = export_folder / "inc5000_analysis.xlsx"

with pd.ExcelWriter(
    output_file,
    engine="openpyxl"
) as writer:

    kpis.to_excel(
        writer,
        sheet_name="Executive KPIs",
        index=False
    )

    dataset_summary.to_excel(
        writer,
        sheet_name="Dataset Summary",
        index=False
    )

    industry_summary.to_excel(
        writer,
        sheet_name="Industry Summary",
        index=False
    )

    state_summary.to_excel(
        writer,
        sheet_name="State Summary",
        index=False
    )

    metro_summary.to_excel(
        writer,
        sheet_name="Metro Summary",
        index=False
    )

    revenue_category.to_excel(
        writer,
        sheet_name="Revenue Category",
        index=False
    )

    growth_category.to_excel(
        writer,
        sheet_name="Growth Category",
        index=False
    )

    company_size.to_excel(
        writer,
        sheet_name="Company Size",
        index=False
    )

    employee_size.to_excel(
        writer,
        sheet_name="Employee Size",
        index=False
    )

    top_revenue.to_excel(
        writer,
        sheet_name="Top Revenue",
        index=False
    )

    top_growth.to_excel(
        writer,
        sheet_name="Top Growth",
        index=False
    )

    efficiency.to_excel(
        writer,
        sheet_name="Revenue Efficiency",
        index=False
    )

    ranking.to_excel(
        writer,
        sheet_name="Company Ranking",
        index=False
    )

print("Workbook exported successfully!")
print(output_file)
````
